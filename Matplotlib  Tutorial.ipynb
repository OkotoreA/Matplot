```python
import matplotlib.pyplot as plt
import numpy as numpy 
xplot=np.array(['a', 'b','c','d'])
yplot=np.array([1,2,3,4,])
plt.bar(xplot,yplot)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['A','B','C', 'D', 'E'])
y=np.array([1,2,3,4,5])
plt.bar(x, y ,color=["blue","red","Orange", "Purple", "yellow"], width= 0.5,)
plt.show()
```

```python
name=np.array(['Apple', 'Banana', 'Orange'])
Var=np.array([300,250,400])
plt.barh(name, Var, color = ["purple", "indigo", "green"])
plt.show()
```

# Histogram
